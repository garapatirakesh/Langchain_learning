# Agent-to-Agent (A2A) Protocols

## Introduction
The **A2A Protocol** is a standardized way for AI agents to communicate, collaborate, and hand off tasks. 

### MCP vs A2A
- **MCP**: Model connected to Tools.
- **A2A**: Agent connected to Agent.


## 1. Discovery (Agent Cards)
In A2A, an agent describes what it can do in a standardized format.

```python
agent_card = {
    'name': 'PythonExpert',
    'capabilities': ['code_generation'],
    'status': 'available'
}
print(json.dumps(agent_card, indent=2))
```

## 2. Tasking
A consumer agent sends a structured request to the provider.

```python
task_request = {
    'jsonrpc': '2.0',
    'method': 'create_task',
    'params': {
        'id': 'T123',
        'prompt': 'Write a sort function.'
    }
}
print(json.dumps(task_request, indent=2))
```

## 3. Artifacts
The response contains artifacts (the work product).

```python
artifact = {
    'type': 'code',
    'language': 'python',
    'content': 'def sort(x): ...'
}
print(json.dumps(artifact, indent=2))
```

## Summary
A2A enables a 'Swarm' of specialized agents to solve complex problems through standardized discovery and delegation.
